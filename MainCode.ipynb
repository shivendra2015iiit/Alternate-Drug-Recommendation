```python
import pandas as pd
```

```python
provider = pd.read_csv("Provider.csv")
Database = pd.read_csv("Database.csv")
```

```python
Database.head(3)
```

```python
listProvider=list(provider["Medicine Required"])

```

```python
sku=list(Database["SKU/PRODUCT"].values)
molecule = list(Database["MOLECULE"].values) 
```

```python
import difflib
```

```python
mainDF = pd.DataFrame(columns=['MEDICINE REQUIRED','CURRENT COMPANY','MOLECULE','THERAPY','CLASS','SUPER GROUP','GROUP','SUBSTITUTE #1','COMPANY #1','SUBSTITUTE #2','COMPANY #2','SUBSTITUTE #3','COMPANY #3'])
for i in listProvider:
    try:
        #provides the closest product available
        
        med_name=difflib.get_close_matches(i,sku)[0]
        med_data=Database.iloc[sku.index(med_name)]
        
        molecule = med_data.MOLECULE
        currentCompany = med_data.COMPANY
        therapy  = med_data.THERAPY
        Class = med_data.CLASS
        supergroup= med_data["SUPER GROUP"]
        group = med_data["GROUP"]
        
        
        tempDB=Database.loc[Database["MOLECULE"] == molecule]
        closermeds=list(tempDB.loc[Database["COMPANY"] != currentCompany]["SKU/PRODUCT"])
        matched_comp= difflib.get_close_matches(i,closermeds)
        
        try:
            
            
            sc1    =  Database.iloc[sku.index(matched_comp[0])]["COMPANY"]
            sm1    =  matched_comp[0]
            
            try:
                sc2    =  Database.iloc[sku.index(matched_comp[1])]["COMPANY"]
                sm2    =  matched_comp[1]
                
                try:

                    sc3    =  Database.iloc[sku.index(matched_comp[2])]["COMPANY"]
                    sm3    =  matched_comp[2]
                    
                except:
                    sc3 = "NA"
                    sm3 = "NA"
                    
            except:
                sc2 = "NA"
                sm2 = "NA"
                sc3 = "NA"
                sm3 = "NA"
                
                
        except:
            
            sc1 = "NA"
            sm1 = "NA"
            sc2 = "NA"
            sm2 = "NA"
            sc3 = "NA"
            sm3 = "NA"

            
    except:
        #NA as drug isn't available in database
        
        molecule = "NA"
        currentCompany = "NA"
        therapy  = "NA"
        Class = "NA"
        supergroup= "NA"
        group = "NA"
        sc1 = "NA"
        sm1 = "NA"
        sc2 = "NA"
        sm2 = "NA"
        sc3 = "NA"
        sm3 = "NA"
        
        
    mainDF=mainDF.append(pd.DataFrame([[i,currentCompany,molecule,therapy,Class,supergroup,group,sc1,sm1,sc2,sm2,sc3,sm3]],
                              columns=['MEDICINE REQUIRED','CURRENT COMPANY','MOLECULE','THERAPY','CLASS','SUPER GROUP','GROUP'
                              ,'SUBSTITUTE #1','COMPANY #1','SUBSTITUTE #2','COMPANY #2','SUBSTITUTE #3','COMPANY #3']),ignore_index=True) 
    
```
